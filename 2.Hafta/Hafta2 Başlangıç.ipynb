# print() Fonksiyonunun Parametreleri
## 1.sep Parametresi
- sep ifadesi, İngilizcede separator (ayırıcı, ayraç) kelimesinin kısaltmasıdır.
- Dolayısıyla print() fonksiyonundaki bu sep parametresi, ekrana basılacak öğeler arasına hangi karakterin yerleştirileceğini gösterir. 
- Kullanılmasının veya kullanılmamasının pek bir önemi yoktur. Çünkü görünmese bile hep oradadır.

```python
print("Adana", "Mersin", sep="-")
```

#2.end Parametresi
- Tıpkı sep parametresi gibi, end parametresi de print() fonksiyonunda görünmese bile her zaman oradadır. 
- end parametresi ise bu parametrelerin sonuna neyin geleceğini belirler.

```python
print("Bugün günlerden Salı", end=".")
```
